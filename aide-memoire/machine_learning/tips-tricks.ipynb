## Tips and Tricks 

#### Euclidean distance between two points

```python
import numpy as np

dist = np.linalg.norm(np.array([2,2,2]) - np.array([1,-1,1]))
dist
```

#### Inverse of a matrix

```python
# a = np.matrix(np.ones([3,3]))
np.linalg.inv(np.matrix([1]))
# print(ainv)
```

### Train test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(dist*np.eye(3), np.eye(3), test_size=0.1, random_state=42)
```

### MultiVARIATE : multiple variables (might be independent)

```python
import matplotlib.pyplot as plt
mean = [0,0]; cov = [[1,0], [0,1]]
x, y = np.random.multivariate_normal(mean, cov, 5000).T
plt.plot(x, y, 'x')
plt.axis('equal')
plt.show()
```
