```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-5, 5, .4)


def f(x):
    return pow(x, 2.0) 


def dfx(x):
    return 2.0 * x

def dfx_roots():
    roots = []
    for i in x:
       #print(dfx(i))
       if abs(dfx(i)) < .000001:
           roots.append(i)
    return roots

# plt.figure(1)
plt.plot(x, f(x), 'g')
plt.plot(x, dfx(x), 'c--')




```

```python
for i in np.arange(-5,5,2):
    x_i = np.arange(i - .5, i + .5, .25)
    m = dfx(i)
    c =  f(i) - m*i
    y_i = m*(x_i)  +  c
    plt.plot(x_i,y_i,'b')

roots = dfx_roots()
for i in roots:
    x_i = np.arange(i - 1, i + 1, .5)
    m = dfx(i)
    c = f(i) - m * i
    y_i = m * (x_i) + c
    plt.plot(x_i, y_i, 'k')

plt.plot(.5,f(.5),'k*')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gradient Descent Parabola')
plt.grid(axis='both',color='c', alpha=0.25)
plt.show()
```
