```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(-2*np.pi,2*np.pi,.05)
# x
```

```python
f_x = np.cos(x)
```

```python
fig = plt.figure()
ax = plt.axes()
plt.grid(axis='both')
plt.plot(x,f_x,'b-')
plt.show()
```

```python
x_t = np.arange(-2,2,.25)
t_f_x = 1 - .5*x_t**2 + (1/24)*x_t**4
```

```python
fig = plt.figure()
ax = plt.axes()
plt.grid(axis='both')
plt.plot(x,f_x,'b-')
plt.plot(x_t, t_f_x,'g*')
plt.plot(3.14 + x_t, -t_f_x,'g*')
plt.plot(-3.14 + x_t, -t_f_x,'g*')
plt.show()
```
