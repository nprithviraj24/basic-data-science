# Binomial Distribution

This type of distribution is applied when there are only two possible outcomesm success or failure.
A random variable X that has a binomial distribution represents the number of successes in a sequence of n independent yes/no trials, each of which yields success with probability p.

Example: A student attempts 10 questions in an exam and the student  has not prepared for the exam. The exam is of multiple choice questions with 4 possible answers. P_Success(6).


```python
import math
from math import factorial
import scipy.stats
from scipy.stats import binom
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline  

n = 10
s = 6
f = 4
ps = .25
pf = .75
#number of ways a student can get 6 correct out of 10 questions is 10C4
combination = factorial(n)/(factorial(s)*factorial(f))
ps6 = ((ps**s)*(pf**f))*combination
print(ps6)
print("%.3e" % ps6)

psAllPosibilities = binom.pmf(s,n, ps)
print(ps6)

#for success from 0 to 10
s = range(11)
ps1_10 = binom.pmf(s,n, ps)
print(ps1_10)

plt.plot(s, ps1_10,'o-')
plt.title('Bionomial distribution')
plt.xlabel('Number of successes')
plt.ylabel('Probability of success')
plt.show()
```

# Poisson distribution
A Poisson distribution is a distribution which shows the likely number of times that an event will occur within a pre-determined period of time. It is used for independent events which occur at a constant rate within a given interval of time. The Poisson distribution is a discrete function, meaning that the event can only be measured as occurring or not as occurring, meaning the variable can only be measured in whole numbers.

Example:
Number of cars passing through an intersection at an given hour. Here we need to estimate count of cars which can be 0 to any number in a fixed time interval assuming that other conditions like rain, rush hour etc remain same. Also the number of cars passing through predetermined intervals are independen. First step is to estimate the expected value $\lambda$ of number of cars passing through an intersection at a given hour.  



```python
# Possion distribution hold for mean of maximum 12
mu = [5,9,12,15]
line = ['b--','b-','g-','g--' ]
mu_x = np.array([])
exp_mu = np.array([])
fact_x = np.array([])
ratio = np.array([])
Y = np.array([])

def p_x(mu, X):   
    global mu_x
    global exp_mu
    global fact_x
    global ratio
    Y = np.array([])
    
    
    for x in X:
        exp_mu = np.append(exp_mu, np.exp(-mu))
        mu_x = np.append(mu_x, np.power(mu,x))
        fact_x = np.append(fact_x, math.factorial(x))    
        ratio = np.append(ratio, np.power(mu,x)/math.factorial(x))       
        
        dist_value =  (np.power(mu,x)*np.exp(-mu))/math.factorial(x)
        #if dist_value < 0.0:
        #    dist_value = 0.0  
        Y  = np.append(Y,dist_value)
    
    return Y
   
def p_x_poisson(mu, X):   
    global mu_x
    global exp_mu
    global fact_x
    global ratio
    Y = np.array([])   
    
    for x in X:
        exp_mu = np.exp(-mu)
        mu_x =  np.power(mu,x)
        fact_x =  math.factorial(x)
        ratio = np.power(mu,x)/math.factorial(x)          
    return Y

#print(p_x(mu,10))

X = np.arange(0,25)


for index, mean in enumerate(mu):
    Y = p_x(mean, X)    
    y_7 = p_x(mean,np.array([7]))    
    X_Less_7 = np.arange(0,8)
    Y_Less_7 = p_x(mean, X_Less_7)    
    plt.plot(7, y_7,'go')
    plt.fill_between(X_Less_7, Y_Less_7)
    plt.plot(X,Y,line[index], label='mu = {}, p(7) = {}'.format(mean, np.round(np.sum(Y_Less_7),2)))     

plt.legend()
plt.show()
```

# Exponential distribution

is the opposite of Piosson distribution
Piosson distribution is estimation of number of events in given interval of time
It estimates time taken between occurance between two events.
In Exponential distribution has parameter lambda which is average number of events
in 1 unit of time.

The probability that X (time between two events) is larger than a perticular time
x is exp(-lambda.x)
The probability that X (time between two events) is less than a perticular time
x is  (1- exp(-lambda.x))

```python
lamb = 10/60
 

# Function for the CDF of the exponential distribution
def CDFExponential(lamb,x): #lamb = lambda
    if x<=0:
        cdf=0
    else:
        cdf=1-np.exp(-lamb*x)
    return cdf

def LessThanLambdaExponential(lamb,x):
    return 1-np.exp(-lamb*x)

def MoreThanLambdaExponential(lamb,x):
    return 1-np.exp(-lamb*x)

#Function to compute the mean of the exponential distribution
def MeanExponential(lamb):
    return 1/lamb;

#Function to compute the variance of the exponential distribution
def VarianceExponential(lamb):
    return (1/lamb)**2;



#CDF of the exponential distribution
#when lambda=1/5 and x=3




#Generate 1000 random random samples
#from a standard uniform distribution
#U=np.random.uniform(0,1,1000)
#lamb=10/60
#X=-np.log(1-U)/lamb
X = np.random.randint(1,10,100)    
X = np.sort(X)





#Exponential CDF plot
#plt.hist(X)
#plt.title("Exponential Histogram")
#plt.xlabel("Value")
#plt.ylabel("Frequency")
#plt.show()


#Exponential pdf histogram
n, bins, patches = plt.hist(X, 10, normed=1, facecolor='blue', alpha=0.50)
#Add best fit line
from scipy.stats import expon
rv = expon()
lamb = plt.plot(bins, lamb*rv.pdf(bins*lamb), 'r--', linewidth=1)

plt.xlabel('Value')
plt.ylabel('Probability')
#plt.title("Uniform Histogram and its PDF")
plt.axis([0, 50, 0, 0.3])
plt.grid(True)
#plt.show()



#print ('Exponential distribution with lambda is 10/60 and x = 3 ', CDFExponential(lamb,3))
#print ('Mean of the distribution with above distribution ', MeanExponential(lamb))
#print ('Variance of the above distribution', VarianceExponential(lamb))
```

# Normal distribution

The normal distribution is a continuous distribution or a function that can take on values anywhere on the real line.
The normal distribution is parameterized by two parameters: the mean of the distribution $\mu$ and the variance $\sigma^2$.

<img src="NormalDistributionEquation.png"> </img>

Mean

Mean, $\bar{X}$, is the average of all the values in a sample. It can be calculated by adding up all the values and then dividing the sum total by the number of values in the sample. When the provided list represents a statistical population, then the mean is called the population mean, $\mu$.




```python
import numpy as np
from scipy.stats import norm
#------------------------------------------------------------
# Define the distribution parameters
sigma_values = [0.5, 1.0, 2.0]
mu = 0
#------------------------------------------------------------
# plot the distributions
linestyles = ['-', '--', ':']
x = np.linspace(-10, 10, 1000)
fig, ax = plt.subplots(figsize=(10, 5))

for sigma, ls in zip(sigma_values, linestyles):
    # create a gaussian / normal distribution
    dist = norm(mu, sigma)

    plt.plot(x, dist.pdf(x), ls=ls, c='black',
             label=r'$\mu=%i,\ \sigma=%.1f$' % (mu, sigma))

plt.xlim(-5.0, 5.0)
plt.ylim(0, 1.0)

plt.xlabel('$x$')
plt.ylabel(r'$p(x|\mu,\sigma)$')
plt.title('Gaussian Distribution')

plt.legend()
plt.show()
```
