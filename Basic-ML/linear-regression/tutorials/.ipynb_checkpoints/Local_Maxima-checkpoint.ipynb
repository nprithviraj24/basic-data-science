```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(-3,3,.1)
print(x)
```

```python
def function_a(var):
    polynomial = -var**2
    return polynomial
    
```

```python
f_x = function_a(x)
```

```python
fig = plt.figure()
ax = plt.axes()
plt.grid(axis='both')
plt.plot(x,f_x,'b-')
plt.show()
```

```python
def f_x_der(x):
    return 2*x
    
```

```python
first_der = f_x_der(x)
fig = plt.figure()
ax = plt.axes()
plt.grid(axis='both')
plt.plot(x,f_x,'b-')
#b-   :  indicates that line must be normal line with blue in colour 
plt.plot(x, first_der, 'c--')
plt.show()
```

```python
def f_x_roots(x):
    roots = []
    #print(type(roots))
    for i in x:
        if (abs(f_x_der(i)) < .000000000001):
           
            roots.append(i)
    return roots
```

```python
roots_array = f_x_roots(x)
print(roots_array)
```
