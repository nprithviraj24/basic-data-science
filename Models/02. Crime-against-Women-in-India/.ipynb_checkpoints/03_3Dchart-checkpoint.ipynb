```python

pop02to10 = pd.DataFrame(columns=[ 'India/State/Union Territory', 'Total - 2002', 'Total - 2003', 'Total - 2004', 'Total - 2005', 'Total - 2006', 'Total - 2007', 'Total - 2008', 'Total - 2009', 'Total - 2010'])

for index, row in pop01and11.iterrows():
    x = int(row[2]) - int(row[1])
    x=int(x/10)
#     print(x)
   
    pop02to10 = pop02to10.append({ 'India/State/Union Territory' : row[0], 'Total - 2002' : row[1] + x, 'Total - 2003' : row[1] + 2*x, 'Total - 2004' : row[1] + 3*x, 'Total - 2005' : row[1] + 4*x, 'Total - 2006' : row[1] + 5*x, 'Total - 2007' : row[1] + 6*x, 'Total - 2008' : row[1] + 7*x, 'Total - 2009' : row[1] + 8*x, 'Total - 2010' : row[1] + 9*x }, ignore_index =True)
    
#     print(index)
# pop02to10
```

```python
pop = pd.merge(left=pop01and11, right=pop02to10, how='inner', left_on='India/State/Union Territory', right_on='India/State/Union Territory')
pop = pop[['India/State/Union Territory', 'Total - 2001', 'Total - 2002','Total - 2003','Total - 2004','Total - 2005','Total - 2006','Total - 2007','Total - 2008', 'Total - 2009', 'Total - 2010', 'Total - 2011']]
pop = pop.rename(index=str, columns={"Total - 2001" : "2001", "Total - 2002": "2002", "Total - 2003": "2003", "Total - 2004": "2004", "Total - 2005": "2005", "Total - 2006": "2006", "Total - 2007": "2007", "Total - 2008": "2008", "Total - 2009": "2009", "Total - 2010": "2010", "Total - 2011": "2011"  })
pop.to_csv('data/pop.csv')
```

```python
graph = pd.DataFrame(columns=['State/UT' , 'Year', 'Population', 'Part', 'Rape', 'Kidnap'])

states = ['ANDHRA PRADESH', 'TAMIL NADU', 'GUJARAT', 'MAHARASHTRA', 'BIHAR', 'MIZORAM', 'WEST BENGAL', 'MADHYA PRADESH', 'UTTAR PRADESH', 'HARYANA', 'PUNJAB' ]
parts = ['SOUTH', 'SOUTH', 'WEST', 'WEST', 'EAST', 'EAST', 'EAST', 'CENTRAL', 'CENTRAL', 'NORTH', 'NORTH']
selected_years = [2001, 2003, 2005, 2007, 2009, 2011]

for i in states:
    for j in selected_years:
        row1 = yrs.loc[(yrs['STATE/UT'] == i) & (yrs['Year'] == j)]
        row2 = pop.loc[(pop['India/State/Union Territory'] == i)]
#         print(row2[str(j)].values)
        graph = graph.append({ 'State/UT': i, 'Year': j, 'Population': row2[str(j)].values, 'Part': parts[states.index(str(i))], 'Rape': row1['Rape'].values, 'Kidnap': row1['Kidnapping and Abduction'].values  }, ignore_index=True )

graph['Population'] = graph['Population'].str.get(0)
graph['Rape'] = graph['Rape'].str.get(0)
graph['Kidnap'] = graph['Kidnap'].str.get(0)
graph.sample(5)
```

```python
import chart_studio.plotly as py
import plotly.graph_objs as go

import pandas as pd

# Get Data: this ex will only use part of it (i.e. rows 750-1500)
# df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv')
df = graph
trace1 = go.Scatter3d(
    x=df['Year'][1:65],
    y=df['Part'][1:65],
    z=df['Population'][1:65],
    text=df['State/UT'][1:65],
    mode='markers',
    marker=dict(
        sizemode='diameter',
        sizeref=69,
        size=df['Rape'][1:65],
        color = df['Kidnap'][1:65],
        colorscale = 'Viridis',
        colorbar = dict(title = 'Kidnap'),
        line=dict(color='rgb(140, 140, 170)')
    )
)

data=[trace1]

layout=go.Layout(height=800, width=800, title='Examining Rape and Kidnap Over Time')

fig=go.Figure(data=data, layout=layout)
py.iplot(fig, filename='3DBubble')
```

```python
import chart_studio.plotly as py
import pandas as pd

# The datasets' url. Thanks Jennifer Bryan!
# url_csv = 'http://www.stat.ubc.ca/~jenny/notOcto/STAT545A/examples/gapminder/data/gapminderDataFiveYear.txt'

# df = pd.read_csv(url_csv, sep='\t')
# df.head()




count = ['ANDHRA PRADESH', 'UTTAR PRADESH', 'HARYANA', 'BIHAR', 'WEST BENGAL']
fill_colors = ['#66c2a5', '#fc8d62', '#8da0cb', '#e78ac3', '#a6d854']
gf = df.groupby('State/UT')

data = []
```

```python
from IPython.display import display, HTML

display(HTML(data="""
<style>
    div#notebook-container    { width: 95%; }
    div#menubar-container     { height: 100%; }
    div#maintoolbar-container { width: 99%; }
    
</style>
"""))
```

```python
for cou, fill_color in zip(count[::-1], fill_colors):
    group = gf.get_group(cou)
    years = group['Year'].tolist()
    length = len(years)
    cou_coords = [cou] * length
    pop = group['Kidnap'].tolist()
    zeros = [0] * length
    
    data.append(dict(
        type='scatter3d',
        mode='lines',
        x=years + years[::-1] + [years[0]],  # year loop: in incr. order then in decr. order then years[0]
        y=cou_coords * 2 + [cou_coords[0]],
        z=pop + zeros + [pop[0]],
        name='',
        surfaceaxis=1, # add a surface axis ('1' refers to axes[1] i.e. the y-axis)
        surfacecolor=fill_color,
        line=dict(
            color='black',
            width=4
        ),
    ))

layout = dict(
    title='Kidnap cases from 2002 to 2010 [Gapminder]',
    showlegend=False,
    scene=dict(
        xaxis=dict(title=''),
        yaxis=dict(title=''),
        zaxis=dict(title=''),
        camera=dict(
            eye=dict(x=-1.7, y=-1.7, z=0.5)
        )
    )
)

fig = dict(data=data, layout=layout)

# IPython notebook
# py.iplot(fig, filename='filled-3d-lines')

py.iplot(fig, filename='filled-3d-lines')
```
